```python
import pandas as pd
import re

# Load the Excel file
path = "/Users/bhuman/development/flagged_kirtan_data.xlsx"
df = pd.read_excel(path, engine='openpyxl')

# Step 1: Clean CATEGORY column (remove trailing/leading spaces)
df['CATEGORY'] = df['CATEGORY'].astype(str).str.strip()

# Step 2: Filter rows where CATEGORY is 'CHANDAN VAGHA'
filtered_df = df[df['CATEGORY'] == 'CHANDAN VAGHA']

# Step 3: Extract and clean 'ON WHICH IT' entries (remove leading numbers, spaces)
def clean_title(text):
    if pd.isna(text):
        return ''
    return re.sub(r'^\d+\s*', '', str(text).strip())

on_which_it_cleaned = set(clean_title(val) for val in filtered_df['ON WHICH IT'].dropna())

# Step 4: Clean 'SONGS' column similarly
df['SONGS_CLEAN'] = df['SONGS'].apply(clean_title)

# Step 5: Keep only rows where cleaned 'SONGS' is NOT in cleaned ON WHICH IT list
non_matching_df = df[~df['SONGS_CLEAN'].isin(on_which_it_cleaned)]

# Step 6: Drop the helper column before saving
non_matching_df = non_matching_df.drop(columns=['SONGS_CLEAN'])

# Step 7: Save the result
non_matching_df.to_excel("filtered.xlsx", index=False)

output_path
```
